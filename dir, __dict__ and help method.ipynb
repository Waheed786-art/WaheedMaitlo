```python


x = [1 , 2 ,3]
print(dir(x))  # dir()  will tell us every attributes of list/tuple/set/dictionary like reverse, sort, reduce , len and many more.
print(x.__add__)

class Student:
    def __init__(self, Name , Age):
        self.Name = Name
        self.Age = Age
    
a = Student("Waheed" , 16)
print(a.__dict__)  # dict function is used to get attributes of a class or constructor.


print(help(Student))  # help function tells us everything about the class or constructor or any other things like string, integer, boolean, float or double float.


y = ["Python", "C++" , "Javascript"]
print(dir(y))

class Teacher:
    def __init__(self, subject , name):
        self.name = name
        self.subject = subject
    
t = Teacher("Chemistry" , "Ghafoor")
print(t.__dict__)
print(help(t))
```

The dir() method:
dir(): The dir() function returns a list of all the attributes and methods (including dunder methods) available for an object. It is a useful tool for discovering what you can do with an object.
>>> x = [1, 2, 3]
>>> dir(x)

The __dict__ attribute
__dict__: The __dict__ attribute returns a dictionary representation of an object's attributes. It is a useful tool for introspection. Example:
>>> class Person:
...     def __init__(self, name, age):
...         self.name = name
...         self.age = age
...
>>> p = Person("John", 30)
>>> p.__dict__

The help() method:
help(): The help() function is used to get help documentation for an object, including a description of its attributes and methods. Example:
>>> print(help(string))

 In conclusion, dir(), dict, and help() are useful built-in functions in Python that can be used to get information about objects. They are valuable tools for introspection and discovery.
